# 2) Drop unqualified variables

*   Drop variables with missing > 50%
*   Drop categorical variables with flat values > 70% (variables with the same value in the same column)

<font color='blue'>Q2) How many columns do we have left?</font>




```python
nomissing = df.dropna(thresh=df.shape[0]/2, axis=1)

nomissing.info()
```

```python
noflat = nomissing.copy()
for col in noflat.columns:
  if noflat[col].value_counts().iloc[0] > 0.7 * noflat.shape[0]:
    noflat = noflat.drop(col, axis=1)

noflat.info()
```

# 3) Remove all rows with missing target (the variable "Survived")

<font color='blue'>Q3) How many rows do we have left?</font>

```python
noMissingTarget = noflat.dropna(subset=["Survived"])

noMissingTarget = noMissingTarget.reset_index().drop("index", axis=1)

noMissingTarget
```

# 4) Handle outliers

For the variable â€œFareâ€, replace outlier values with the boundary values


*   If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
*   If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)

<font color='blue'>Q4) What is the mean of â€œFareâ€ after replacing the outliers (round 2 decimal points)?</font>



```python
Q1, Q3 = noMissingTarget["Fare"].quantile([0.25, 0.75])
leftVal = Q1 - 1.5*(Q3 - Q1)
rightVal = Q3 + 1.5*(Q3 - Q1)

noOutlier = noMissingTarget.copy()
noOutlier["Fare"] = np.clip(noOutlier["Fare"], leftVal, rightVal)

noOutlier.info()
```

# 5) Impute missing value



*   Show the number of missing values in each variable
*   Impute missing values with mean & mode
*   Show the number of missing values again after missing value imputation

<font color='blue'>Q5) Which variable has the largest number of missing values?</font>



```python
noOutlier.head()
```

```python
from sklearn.impute import SimpleImputer

imputed = noOutlier.copy()

print(imputed.isnull().sum().idxmax())

num_imp=SimpleImputer(strategy='mean')
cat_imp=SimpleImputer(strategy='most_frequent')
numCol = ['Age', 'Fare']
catCol = ['Pclass', 'Name', 'Sex', 'SibSp', 'Ticket', 'Embarked']

imputed[numCol] = pd.DataFrame(num_imp.fit_transform(imputed[numCol]), index=imputed.index)
imputed[catCol] = pd.DataFrame(cat_imp.fit_transform(imputed[catCol]), index=imputed.index)

imputed.info()
```

# 6) Convert categorical to numeric values



*   Drop the variables â€œNameâ€ & â€œTicketâ€
*   For the variables â€œSexâ€ & â€œEmbarkedâ€, perform the dummy coding and drop the first level. Also, drop those original variables (â€œSexâ€ & â€œEmbarkedâ€)

<font color='blue'>Q6) How many columns do we have?</font>



```python
converted = imputed.copy().drop(["Name", "Ticket"], axis=1)

from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore', drop='first') # this feature will be all zeros
# passing bridge-types-cat column (label encoded values of bridge_types)
nominal_columns = ["Sex", "Embarked"]
enc_df = pd.DataFrame(enc.fit_transform(converted[nominal_columns]).toarray(), columns=enc.get_feature_names_out())

encoded = pd.concat([converted, enc_df], axis=1)
encoded.drop(nominal_columns, axis=1, inplace=True)
encoded
```

# 7) Partition data

*   Show a proportion between survived (1) and died (0) in all data sets (total data, train, test)
