```python
videos = pd.read_csv('data/GBvideos.csv')

videos
```

## How many rows are there in the GBvideos.csv after removing duplications?

```python
videos = videos.drop_duplicates()

videos.info()
```

## How many VDO's that contain at least one record (row) of "dislikes" more than "likes"?  <font color=red>DO NOT group by the data and make sure that you count only unique title!</font>

```python
dislikeMorethanLike = videos[videos["dislikes"] > videos["likes"]].drop_duplicates(["title"])

dislikeMorethanLike.info()
```

## How many VDO that are trending on 22 Jan 2018 with comments more than 10,000 comments?

```python
videos[(videos["trending_date"] == '18.22.01') & (videos["comment_count"] > 10000)].info()
```

## Which date that has the minimum average number of comments per VDO?

```python
avgComment = videos.groupby("trending_date").mean("comment_count")
avgComment.idxmin()["comment_count"]
```

## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
videos
```

```python
category = pd.read_json('data/GB_category_id.json')

categoryDict = {}
for c in category["items"]:
  categoryDict[int(c['id'])] = c['snippet']['title']

print(categoryDict)
```

```python
cat_videos = videos.replace({"category_id": categoryDict})

sports = cat_videos[cat_videos['category_id'] == 'Sports'].groupby("trending_date").sum().rename(columns={"views": "Sports"})[["Sports"]].reset_index()
comedy = cat_videos[cat_videos['category_id'] == 'Comedy'].groupby("trending_date").sum().rename(columns={"views": "Comedy"})[["Comedy"]].reset_index()

sports
```

```python
merged = sports.merge(comedy)

comp = merged[merged["Sports"] > merged["Comedy"]]

comp.info()
```
