```python
import pandas as pd
import numpy as np
```

```python
filePath = 'mushroom2020_dataset.csv'
df = pd.read_csv(filePath)

df.info()
```

```python
# drop row where the target viable is missing
noMissing = df.dropna(subset=["label"]).reset_index().drop('index', axis=1)

noMissing.info()
```

```python
dropColumns = ['id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate','stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type']

onlyUse = noMissing.drop(dropColumns, axis=1)

onlyUse.info()
```

```python
onlyUse.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

numCol = ['cap-color-rate']
noCol = onlyUse.columns.to_list()
noCol.remove('cap-color-rate')
noCol.remove('label')

numImputer = SimpleImputer(strategy='mean')
noImputer = SimpleImputer(strategy='most_frequent')

imputed = onlyUse.copy()
imputed[numCol] = pd.DataFrame(numImputer.fit_transform(imputed[numCol]))
imputed[noCol] = pd.DataFrame(noImputer.fit_transform(imputed[noCol]))

imputed.info()
```

```python
dummies = imputed.copy()

labelDict = {
    "e": 1,
    "p": 0
}

dummies = imputed.replace({"label": labelDict})

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(drop='first', handle_unknown='ignore')
encoded_df = pd.DataFrame(enc.fit_transform(dummies[noCol]).toarray(), columns=enc.get_feature_names_out())
dummies.drop(noCol, axis=1, inplace=True)
dummies = pd.concat([dummies, encoded_df], axis=1)

dummies.info()
```

```python
from sklearn.model_selection import train_test_split

X = dummies.drop('label', axis=1)
y = dummies['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2020)

X_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

options = {
    'criterion':['gini','entropy'],
    'max_depth': [2,3,6],
    'min_samples_leaf':[2,5,10],
    'n_estimators':[100,200],
    'random_state': [2020]
}

gridSearch = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=options,
    cv=5,
    n_jobs=-1
)

gridSearch.fit(X_train, y_train)

bestModel = gridSearch.best_estimator_
```

```python
bestModel
```

```python
from sklearn.metrics import classification_report,confusion_matrix

predictions = bestModel.predict(X_test)

print(classification_report(y_test, predictions, digits=4))
print(confusion_matrix(y_test, predictions, labels=[0, 1]))
```
